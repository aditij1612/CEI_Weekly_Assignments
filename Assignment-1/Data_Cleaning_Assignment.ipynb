Handle missing values

```python
df.isnull().sum()
```

To handle missing values we can either drop them or fillup them

```python
# Fill text columns with "NA"
text_columns = df.select_dtypes(include=['object']).columns

for col in text_columns:
    df[col] = df[col].fillna("NA")


# Drop any remaining missing values
df = df.dropna()
df
```

Basic operations



```python
df[['title', 'final_price', 'ratings_count']]
```

```python
df[df['rating'] > 4]
```

Remove duplicates

```python
df = df.drop_duplicates()
df
```

Create a derived column

```python
df['final_price'] = df['initial_price'] - (
    df['initial_price'] * df['discount'] / 100
)

# Convert final_price into integer
df['final_price'] = df['final_price'].astype(int)
```

```python
df['total_amount'] = df['final_price'] * df['ratings_count']
df
```

```python
df[['title', 'final_price', 'ratings_count', 'total_amount']]
```

Save the cleaned dataset

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
